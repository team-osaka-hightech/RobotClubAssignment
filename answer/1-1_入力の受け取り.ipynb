```python
# 1.
# 参考→1-1-3項

n = int(input())
num_array = []
for i in range(n):
    num_array.append(int(input()))

print(num_array)
```

```python
# 2.
# 参考1-1-4項
# mは引っ掛け。使いません。

n, m = int(input()), int(input())
data_array = []
for i in range(n):
    data_array.append(list(input().split()))

print(data_array)
```

```python
# 3.
# あまりイケてないコードです

num_array = input().split()
for i in range(len(num_array)): # len()はリストの長さを返す
    num_array[i] = int(num_array[i])

print(num_array)
```

```python
# 4.
# 便利なので丸ごと覚えておくとGood
# 参考 https://kakedashi-engineer.appspot.com/2020/02/22/input-list/

num_array = list(map(int, input().split()))

print(num_array)
```
